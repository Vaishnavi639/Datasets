Data cleaning


```python
da = da.fillna(0)
```

```python
cleaned_data = da.drop_duplicates(subset='Company_name', keep='first')
```

```python
cleaned_data.isnull().sum()
```

Central Tendency

```python
print("Mean Rating of Companies = ",cleaned_data['Rating'].mean())
print("Median Rating of Companies = ",cleaned_data['Rating'].median())
print("Mode Rating of Companies = ",cleaned_data['Rating'].mode())
```

```python
print("Mean Avg_salary of Companies = ",cleaned_data['Avg_salary'].mean())
print("Median Avg_salary of Companies = ",cleaned_data['Avg_salary'].median())
print("Mode Avg_salary of Companies = ",cleaned_data['Avg_salary'].mode())
print("Max Avg_salary of Companies = ",cleaned_data['Avg_salary'].max())
```

```python
print("Mean Interviews_taken of Companies = ",cleaned_data['Interviews_taken'].mean())
print("Median Interviews_taken of Companies = ",cleaned_data['Interviews_taken'].median())
print("Mode Interviews_taken of Companies = ",cleaned_data['Interviews_taken'].mode())
print("Max Interviews_taken by a company = ",cleaned_data['Interviews_taken'].max())
```

Create a sample

```python
sample1 = cleaned_data.sample(250)
sample1.head(5) 
```

Distribution of Rating
 

```python
cleaned_data.dropna(subset=['Rating'], inplace=True)
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['Rating'], bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

Scatterplot on the basis of Average Salary and Total Reviews

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Avg_salary', y='Total_reviews', data=sample1)
plt.title('Average Salary and Total Reviews')
plt.xlabel('Total Reviews')
plt.ylabel('Average Salary')
plt.show()
```

Catagories for Highly rated for


```python

plt.figure(figsize=(12, 8))

top_15 = cleaned_data['Highly_rated_for'].value_counts().nlargest(15).index

filtered_data = cleaned_data[cleaned_data['Highly_rated_for'].isin(top_15)]


sns.countplot(y='Highly_rated_for', data=filtered_data, order=top_15)
plt.title('Top 15 Categories in Highly Rated for')
plt.xlabel('Frequency')
plt.ylabel('Categories')
plt.show()
```

Top Companies Rating

```python
top_rated_companies = cleaned_data.nlargest(10, 'Total_reviews')
plt.figure(figsize=(12, 8))
sns.barplot(x='Company_name', y='Total_benefits', data=top_rated_companies, palette='viridis')
plt.title('Top 10 Companies')
plt.xlabel('Companies Name')
plt.ylabel('Total_reviews')
plt.show()
```

Total Benefits of a company using boxplot by a random sample  

```python
 
sample1.boxplot('Total_benefits', grid = False) 
```

```python
bottom_companies = cleaned_data.nsmallest(10, 'Rating')
plt.figure(figsize=(10, 6))
sns.barplot(x='Rating', y='Company_name', data=bottom_companies, palette='inferno')
plt.title('Top 10 Companies with the Worst Ratings')
plt.xlabel('Ratings')
plt.ylabel('Companies Name')
plt.show()
```
